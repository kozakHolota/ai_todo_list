# Введіть API ключ

```python
from llm import get_llm
from tools.util import print_chunk
from time import sleep

openai_api_key = input("Введіть ключ")
llm_agent = get_llm(openai_api_key)
```

# Тести агента
## Автотести

```python


tests = [
    "Додай: купити хліб (через 2 години)",
    "Додай: подзвонити лікарю (до завтра)",
    "Покажи всі завдання",
    "Видали завдання про хліб",
    "Що залишилось?"
]

for test in tests:
    print(f"\n>>> {test}")
    inputs = {"messages": [{"role": "user", "content": test}]}
    for chunk in llm_agent.stream(inputs, stream_mode="updates"):
        print_chunk(chunk)
        sleep(1)
```

## Програмний інтерфейс

```python
request = {"messages": [{"role": "user", "content": input("Введи запит по керуванню твоїми завданнями")}]}

for chunk in llm_agent.stream(request, stream_mode="updates"):
        print_chunk(chunk)
        sleep(1)
```
